```python
import folium
import datetime as dt
import random as rnd
import csv

t0 = dt.datetime.now()

#New York City Coordinates
NYC_COORD = [40.7128, -74.0059]

# Build map 
map_nyc = folium.Map(location=NYC_COORD, zoom_start=13, 
tiles='cartodbpositron', width=640, height=480)
```

```python
map_nyc
```

```python
df=[]

with open('out.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader)
    next(reader)
    next(reader)
    for row in reader:
        df.append(row)
```

```python
for entry in df:
    if entry[1] != "undefined":
        folium.Marker( 
            location = [ float(entry[2]), float(entry[1]) ],
            popup = entry[0]
        ).add_to( map_nyc )
```
